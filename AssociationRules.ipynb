# Data source:

```r
download.file('http://staff.ii.pw.edu.pl/~gprotazi/dydaktyka/dane/supermarket.csv','supermarket.csv')
marketSet = read.csv('supermarket.csv',sep=';')
marketSet= as.data.frame(sapply(marketSet, function(x) as.logical(x)))
summary(marketSet)
head(marketSet)
View(marketSet)
```

Data is divided into product categories and departments. Each record specifies basket from one department. Data is logical so each record through true and false values express one basket.

Convertion from "data.frame" datatype to "transactions" datatype. I need to transform marketSet into transactions data in order to perform operations on it.


```r
marketSet <- as(marketSet, "transactions")
```

## Rule 1

First I will try to find association rules with a confidence equal to 1. I suppose the support with such a big confidence may be very small, that is why I will set the support to 0.015


```r
aParam  = new("APparameter", "confidence" =1, "support" =0.015, "minlen"= 2, target = "rules", maxtime =40 , maxlen = 20) 
```

```r
print(aParam)
```

```r
aRules <-apriori(marketSet,aParam)
```

```r
summary(aRules)
```

A lot of rules have been found however maximum support is only 2,1%. I will investigate  further the findings to see if there is anything that may be advised to the shop owner  


```r
#example of rules
inspect(head(sort(aRules, by="support"),50))
```

Majority of the biggest support rules cover the situation where the antecedent is a relatively big  set of random product and the consequent is {vegetables}, {frozen.foods} or {bread.and.cake}


```r
subsetRHS <- subset(aRules, subset = rhs %in% "vegetables"  |  rhs %in% "frozen.foods" |  rhs %in% "bread.and.cake" )
summary(subsetRHS)
```

```r
length(aRules)
```

Rest of the product as said previously may be hard to connect with the previous.

From the first analysis of association rules it can be seen that {vegetables}, {frozen.foods}, {bread.and.cake}, {biscuits} and {fruits} are often bought when size of the basket is big. Also, lift for those products is higher than 1 meaning that s that the presence of the antecedent of a rule in a given transaction increases the probability of occurrence of the consequent of a rule in that transaction. It can be also understood in this situation as presence of consequent in the transaction increase probability of e antecedent. 

Advise for the shop would be to connect vegetables with fruits and bread and cake with biscuit and put them together with frozen foods close to the beginning of the shop. Customers who would buy one of those products have a bigger chance of doing big shopping.

## Rule 3

Following this track I will investigate further for this rule confirmation by lowering confidence and exceeding support.

```r
aParam@confidence = 0.8
aParam@support = 0.2
aRules <-apriori(marketSet,aParam)
```

```r
summary(aRules)
```

```r
inspect(sort(aRules, by="support"))
```
